```python
# from row 3, with 2 points define a central line
X1 = 322
Y1 = -1079
X2 = 1917
Y2 = -279

m = (Y1 - Y2) / (X1 - X2)
b = Y1 - m * X1

x_values = np.linspace(-10, 2000, 100)
y_values = m * x_values + b

# Define a function to plot traffic condition for a given set of test points and time
def plot_traffic_condition(test_points, date, time):
    # Plot the line
    plt.plot(x_values, y_values, label=f'y = {m:.3f}x + {b:.3f}', color='blue')

    # Loop through each test point to determine its position and plot it
    for x_test, y_test in test_points:
        plt.scatter(x_test, y_test, s=50)

    plt.xlabel('x')
    plt.ylabel('y')
    plt.title(f'Traffic Condition at WCP (2701) on {date} at {time}')
    plt.legend()
    plt.grid(True)
    plt.show()

# Call the function for each subset of test points
plot_traffic_condition(test_points_0349, date_0349, time_0349)
plot_traffic_condition(test_points_0449, date_0449, time_0449)
plot_traffic_condition(test_points_0649, date_0649, time_0649)
plot_traffic_condition(test_points_0849, date_0849, time_0849)
plot_traffic_condition(test_points_1049, date_0649, time_1049)
plot_traffic_condition(test_points_1128, date_1128, time_1128)
plot_traffic_condition(test_points_1208, date_1208, time_1208)
plot_traffic_condition(test_points_1308, date_1308, time_1308)
plot_traffic_condition(test_points_1408, date_1408, time_1408)
plot_traffic_condition(test_points_1508, date_1508, time_1508)
plot_traffic_condition(test_points_1608, date_1608, time_1608)
plot_traffic_condition(test_points_1708, date_1708, time_1708)
plot_traffic_condition(test_points_1809, date_1809, time_1809)
plot_traffic_condition(test_points_1908, date_1908, time_1908)
plot_traffic_condition(test_points_2008, date_2008, time_2008)
plot_traffic_condition(test_points_2108, date_2108, time_2108)
plot_traffic_condition(test_points_2208, date_2208, time_2208)
plot_traffic_condition(test_points_2308, date_2308, time_2308)
```

```python
# to show number of car via plot
time_frames = ["3:49", "4:49", "6:49", "8:49","10:49", "11:28", "12:08", "13:08", 
               "14:08", "15:08", "16:08", "17:08", "18:09", "19:08", "20:08", "21:08", 
               "22:08", "23:08"]
counts = {time: {'JB': 0, 'SG': 0} for time in time_frames}

# Loop through each time frame and count cars going to JB and SG
for time_frame in time_frames:
    df_time = df_12oct[df_12oct['Time'] == time_frame]
    for _, row in df_time.iterrows():
        x, y = row['center_x'], row['center_y_1']
        y_on_line = m * x + b
        if y > y_on_line:
            counts[time_frame]['JB'] += 1
        elif y < y_on_line:
            counts[time_frame]['SG'] += 1

# Convert counts to a DataFrame for easy plotting
counts_df = pd.DataFrame(counts).T  # Transpose to get time frames as rows

# Plotting
fig, ax = plt.subplots(figsize=(12, 6))
counts_df.plot(kind='bar', stacked=False, ax=ax, color=['blue', 'orange'])

# Add numbers on top of each bar
for container in ax.containers:
    ax.bar_label(container, label_type='edge', fontsize=8, padding=3)

# Labeling
ax.set_xlabel('Time')
ax.set_ylabel('Number of Vehicle')
ax.set_title(f'Traffic Condition at WCP (2701) on {date_0349}')
ax.legend(title='Direction')
plt.xticks(rotation=45)
plt.tight_layout()

plt.show()
```
